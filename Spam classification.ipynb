```python
import re
def remove_non_alphabetic(text):
    if isinstance(text, str):
        return re.sub(r'[^a-zA-Z]', ' ', text)
    else:
        return str(text)

# Apply the function to the 'input' column
df['input'] = df['input'].apply(remove_non_alphabetic)

# Display the updated DataFrame
print(df)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
```

```python
import re
for i in range(len(df)):
    try:
        review=re.sub('[^a-zA-Z]',' ',df['input'][i])
        review=review.lower()
        review=review.split()
        stop=set(stopwords.words('english'))
        review=[ps.stem(word) for word in review if  word not in  stop ]
        review=''.join(review)
        corpus.append(review)
    except KeyError as e:
        continue
```

```python
len(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer()
x=vector.fit_transform(corpus).toarray()
```

```python
y=pd.get_dummies(df['output'])
y=y.iloc[:,1]
y.replace({True:1,False:0})
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
mul=MultinomialNB()
mul.fit(x_train,y_train)
y_pred=mul.predict(x_test)
accu=accuracy_score(y_test,y_pred)
accu
```
